```python
import numpy as np
import math
import random
from sklearn.preprocessing import normalize
import matplotlib.pyplot as plt
```

```python
def function1(x):
    x1 = x[0]
    x2 = x[1]
    return x1*x1+x2*x2
```

```python
def function2(x):
    x1 = x[0]
    x2 = x[1]
    return -1*(1+math.cos(12*math.sqrt(pow(x1,2)+pow(x2,2))))/(0.5*(pow(x1,2)+pow(x2,2))+2)
```

```python
def function3(x):
    ans = 0
    for i in range(50):
        ans += pow(x[i],2)      
    return ans
```

### Simulated Annealing

```python
k = 50
func = function3
T = 1000
plot = []
for _ in range(5):
    x = [2]*k
    cov = [[0]*k for i in range(k)]
    for i in range(k):
        cov[i][i]=1
    fi = func(x)
    temp = [fi]
    print("Iteration: "+ str(0)+"  Function Value: "+ str(func(x)))
    for i in range(100):
        Tk = T/(i+1)
        mean = x
        d_x = np.random.multivariate_normal(mean, cov, 1).tolist()[0]

        func_val = func(x)
        func_val_d = func(d_x)

        if(func_val_d<func_val):
            x = d_x
        else:
            prob = math.exp((
                -func_val+func_val_d)/T)
            #print(prob)
            update = random.choices([0,1], weights=(prob,1-prob))
    #         print(update)
            if(update==[1]):
                x = d_x
    #     print(x1)
    #     print(x2)
        f = func(x)
        #if(f>=3*fi): break
        temp.append(f)
        if((i+1)%20==0):
            print("Iteration: "+ str(i+1)+"  Function Value: "+ str(f))
        
    print("**************************************************")
    plot.append(temp)
        #break
    
```

```python
x1 = [i for i in range(len(plot[0]))]
x2 = [i for i in range(len(plot[1]))]
x3 = [i for i in range(len(plot[2]))]
x4 = [i for i in range(len(plot[3]))]
x5 = [i for i in range(len(plot[4]))]


plt.plot(x1, plot[0],label = "Run 1")
plt.plot(x2, plot[1],label = "Run 2")
plt.plot(x3, plot[2],label = "Run 3")
plt.plot(x4, plot[3],label = "Run 4")
plt.plot(x5, plot[4],label = "Run 5")


plt.xlabel("Iterations")
plt.ylabel("Function Value")
#plt.hlines(y=5, xmin=0, xmax=10, colors='purple', linestyles='--', label='Global Minima')
plt.legend(fontsize = 9)

plt.show()
```

### Cross Entropy

```python
func = function3
k = 50
n_samples = 100

plot = []
for _ in range(5):
    temp = []
    mean = [2]*k
    cov = [[0]*k for i in range(k)]
    for i in range(k):
        cov[i][i]=1
    

    f = func(mean)
    temp.append(f)
    print("Iteration: "+ str(0)+"  Function Value: "+ str(f))

    for i in range(100):
        #Sample k points from mean and covariance
        d_x = np.random.multivariate_normal(mean, cov, n_samples).tolist()
        samples = d_x#[[d_x1[i],d_x2[i]] for i in range(k)]
        #print(len(samples))

        #Compute function value for these k points
    #     for i in range(k):
    #         samples[i].append(function(samples[i][0],samples[i][1]))


        #Pick top 40%
        samples = sorted(samples,key=lambda l:func(l))
        samples = samples[:int(len(samples)*0.4)]
        samples = np.array(samples)
        #print(len(samples))
        #print(samples)

        #Recalculate distribution
        mean = samples.mean(axis=0)
        cov = np.cov(samples.T)
        #print(mean)
        #print(cov)

        #Funtion value from mean
        f = func(mean)
        print("Iteration: "+ str(i+1)+"  Function Value: "+ str(f))
        temp.append(f)
        #break
    print("**************************************************")
    plot.append(temp)
    
    
```

### Search Gradient

```python
#theta = np.array([[2,2],[1,0],[0,1]])
function = function3
k = 50
n_samples = 100
lr = 0.01

plot = []
for i in range(5):
    temp = []
    mean = [2]*k
    cov = [[0]*k for i in range(k)]
    for i in range(k):
        cov[i][i]=1
    theta = [mean]
    theta.extend(cov)
    f = function(theta[0])
    temp.append(f)
    print(f)

    for _ in range(100):
        mean = theta[0]
        cov = theta[1:]
        #print(np.linalg.eigvals(cov))
        inv_cov = np.linalg.inv(cov)
        #Draw k samples 
        d_x = np.random.multivariate_normal(mean, cov, n_samples).tolist()
        samples = np.array(d_x)
        #print(samples.shape)

        #Calculate log derievatives and function values
        mean_drv = []
        cov_drv = []
        for j in range(n_samples):
            md = np.matmul(inv_cov,(samples[j]-mean))
            #print(md)
            mean_drv.append(md)

            cd = (-1/2)*inv_cov + (1/2)*(np.matmul(inv_cov,np.matmul(np.matrix(samples[j]-mean).T,np.matmul(np.matrix(samples[j]-mean),inv_cov))))
            #cd = (-1/2)*inv_cov + (1/2)*(np.matmul(inv_cov,np.matmul((samples[i]-mean).T,np.matmul((samples[i]-mean),inv_cov))))
            cov_drv.append(cd)

    #     print(np.sum(mean_drv,axis=0)/k)
    #     print(np.sum(cov_drv,axis=0)/k)

        #Update matrix
        delta = []
        mJ = [function(samples[j])*mean_drv[j] for j in range(n_samples)]
        vJ = [function(samples[j])*cov_drv[j] for j in range(n_samples)]

        mJs = (sum(mJ)/n_samples).tolist()
        vJs = (sum(vJ)/n_samples).tolist()
        #print(vJs)

        delta.append(mJs)
        delta.extend(vJs)
        #print(delta)
        #print(np.linalg.norm(delta))

        #Normalize update matrix
        #delta_norm = normalize(delta, axis=1, norm='l2')
        #print(delta_norm)
        delta_norm = delta/np.linalg.norm(delta)
        #print(delta_norm)


        #Update Step
        theta_u = theta - lr*delta_norm
        #print(theta_u)
        #print(np.linalg.eigvals(theta_u[1:]))

        #Check for PSD
        #if not(np.all(np.linalg.eigvalsh(theta_u[1:])>0-1e-8)): continue

        theta = theta_u
    #     print(i+1)
        #print(theta[0])
        f = function(theta[0])
        temp.append(f)
        print(f)
    print("**************************************************")
    plot.append(temp)

        #break
    
    
```

```python
x1 = [i for i in range(len(plot[0]))]
x2 = [i for i in range(len(plot[1]))]
x3 = [i for i in range(len(plot[2]))]
x4 = [i for i in range(len(plot[3]))]
x5 = [i for i in range(len(plot[4]))]


plt.plot(x1, plot[0],label = "Run 1")
plt.plot(x2, plot[1],label = "Run 2")
plt.plot(x3, plot[2],label = "Run 3")
plt.plot(x4, plot[3],label = "Run 4")
plt.plot(x5, plot[4],label = "Run 5")


plt.xlabel("Iterations")
plt.ylabel("Function Value")
#plt.hlines(y=5, xmin=0, xmax=10, colors='purple', linestyles='--', label='Global Minima')
plt.legend(fontsize = 9)

plt.show()
```
