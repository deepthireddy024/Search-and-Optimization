### Question 4

```python
def function(x1,x2):
    return x1*x1 - x1*x2 + 3*x2*x2 + 5

def gradient(x1,x2):
    norm = math.sqrt(pow(2*x1-x2,2)+pow(6*x2-x1,2))
    return 2*x1-x2 , 6*x2-x1

def hessian(x1,x2):
    h = [[2,-1],[-1,6]]
    return h
```

```python
plot1 = []
alpha = 0.2
x1 = 2
x2 = 2
f = function(x1,x2)
plot1.append(f)
print("Iteration: 0")
print("x1=" + str(round(x1,4))+ "  x2="+ str(round(x2,4)))
print("Function value:"+str(round(f,4)))
print("***********************")

for i in range(10):
    p1,p2 = gradient(x1,x2)
    x1 = x1-alpha*p1
    x2 = x2-alpha*p2
    f = function(x1,x2)
    plot1.append(f)
    print("Iteration: "+str(i+1))
    print("x1=" + str(round(x1,4))+ "  x2="+ str(round(x2,4)))
    print("Function value:"+str(round(f,4)))
    print("***********************")
    
```

```python
plt.plot(plot1)
plt.xlabel("Iterations")
plt.ylabel("Function Value")
plt.show()
```

```python
plot2 = []
alpha = 0.33
x1 = 2
x2 = 2
f = function(x1,x2)
plot2.append(f)
print("Iteration: 0")
print("x1=" + str(round(x1,4))+ "  x2="+ str(round(x2,4)))
print("Function value:"+str(round(f,4)))
print("***********************")

for i in range(10):
    p1,p2 = gradient(x1,x2)
    x1 = x1-alpha*p1
    x2 = x2-alpha*p2
    
    f = function(x1,x2)
    plot2.append(f)
    print("Iteration: "+str(i+1))
    print("x1=" + str(round(x1,4))+ "  x2="+ str(round(x2,4)))
    print("Function value:"+str(round(f,4)))
    print("***********************")
    
```

```python
plt.plot(plot2)
plt.xlabel("Iterations")
plt.ylabel("Function Value")
plt.show()
```

```python
plot3 = []
x1=2
x2=2
f = function(x1,x2)
plot3.append(f)
X = np.matrix([x1,x2]).T
g = np.matrix(gradient(x1,x2)).T
H = np.matrix(hessian(x1,x2))
Xu = X - np.matmul(np.linalg.inv(H),g)
x1 = Xu.tolist()[0][0]
x2 = Xu.tolist()[1][0]
f = function(x1,x2)
plot3.append(f)
```

```python
x = [i for i in range(11)]

plt.plot(x, plot1, label ='1. Gradient Descent Converging(alpha=0.2)')
plt.plot(x, plot2,  label ='2. Gradient Descent Diverging(alpha=0.33)')
plt.plot([0,1], plot3,  label ='3. Newton Descent(alpha=1)')

plt.xlabel("Iterations")
plt.ylabel("Function Value")
plt.hlines(y=5, xmin=0, xmax=10, colors='purple', linestyles='--', label='Global Minima')
plt.legend(fontsize = 9)
plt.xticks(x)
plt.show()
```
