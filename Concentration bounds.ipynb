```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def get_mean(N,trials): 
    #N = 100
    mean = []
    for i in range(trials):
        samples = np.random.choice([0,1],N,p=[0.3,0.7])
        mean.append(np.mean(samples))
    #print(mean)
    return mean
```

```python
bins = []
for i in range(11):
    bins.append(round(i*0.1,2))
bins
```

```python
mean = get_mean(1,500)
plt.hist(mean, bins=bins, edgecolor='black')
plt.xlabel('Mean')
plt.ylabel('Count')
plt.title('N=1')
plt.show()
```

```python
mean = get_mean(50,500)
plt.hist(mean, bins=bins, edgecolor='black')
plt.xlabel('Mean')
plt.ylabel('Count')
plt.title('N=50')
plt.show()
```

```python
def prob(means,act_mean,epsilon):
    cnt = len([i for i in means if abs(i-act_mean)>=epsilon])
    return cnt*(1.0/len(means))
```

```python
x=[]
y=[]
hoeff = []
for N in range(1,101):
    mean = get_mean(N,10)
    ans = prob(mean,0.7,0.1)
    h = math.exp(-2.0*N*0.01)*2.0
    x.append(N)
    y.append(ans)
    hoeff.append(h)
plt.plot(x,y,label=r'$P(|\bar X_N-\mu|\geq \varepsilon)$')
plt.plot(x,hoeff,label='Hoeffding bound')
plt.legend()
plt.show()
```

```python
def prob(means,act_mean,epsilon):
    cnt = len([i for i in means if abs(i-act_mean)>=epsilon])
    return cnt*(1.0/len(means))
```

```python
def get_mean(N,trials): 
    #N = 100
    mean = []
    for i in range(trials):
        samples = np.random.uniform(low=0,high=1,size=N)
        mean.append(np.mean(samples))
    #print(mean)
    return mean
```

```python
def get_z(N,eps):
    mean = get_mean(N,1000)
    #print(mean)
    ans = prob(mean,0.5,eps)
    return ans
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import random

def fun(x, y):
    return get_z(x,y)


x = np.arange(10, 1000, 10)
y = np.arange(0, 0.5, 0.01)
X, Y = np.meshgrid(x, y)
zs = np.array([fun(x,y) for x,y in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.view_init(azim=15, elev=30)

ax.set_xlabel('N')
ax.set_ylabel('epsilon')
ax.set_zlabel('Z Label')

plt.show()
```
