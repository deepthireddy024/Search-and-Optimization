```python
import numpy as np
import matplotlib.pyplot as plot
import pylab
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import math
```

```python
x1 = []
x2 = []
start = -10
step = 0.1
while(1):
    x1.append(start)
    x2.append(start)
    start = start + step
    if(start > 10): break
```

```python
y = np.ndarray((len(x1),len(x2)))
   
for i in range(0, len(x1)):
    for j in range(0, len(x2)):
        y[i][j] = (x1[i]*x2[j])
```

```python
pylab.xlim([-10,10])
pylab.ylim([-10,10])
plot.title('Level sets')
plot.xlabel('x1')
plot.ylabel('x2')

contours = plot.contour(x1, x2, y,levels = 80)
plot.clabel(contours, inline=1, fontsize=5)


plot.show()
```

```python


c1 = 0.0001
k = -2*(np.log(4*pow(math.pi,3/2)*c1))
#print(k)

k1 = math.sqrt(k)

a = k1  
b = math.sqrt(2)*k1  
c = k1  


theta = np.linspace(0, np.pi, 50)         
phi = np.linspace(0, 2 * np.pi, 100)  
Theta, Phi = np.meshgrid(theta, phi)


X = a * np.sin(Theta) * np.cos(Phi)+1
Y = b * np.sin(Theta) * np.sin(Phi)+1
Z = c * np.cos(Theta)+1


fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis')


ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('x3')
ax.zaxis.labelpad = -4.5
ax.axes.set_xlim(-2, 4)
ax.axes.set_ylim(-2, 4)
ax.axes.set_zlim(-2, 4)
ax.set_title('Level Set for c = '+str(c1))

plt.show()
```
